#### 상폐일 존재하지 않는 기업과 존재하는 기업 구분

```python
minor = pd.read_csv("./drive/MyDrive/공유문서함/240312상장_비주력산업(new).csv",index_col=0)
minor
```

```python
minor.reset_index(drop=True, inplace=True)
```

```python
# (주)회사명에서 (주)를 제거_비주력산업
minor.loc[minor['회사명'].str.contains('\(주\)'), '회사명'] = minor.loc[minor['회사명'].str.contains('\(주\)'), '회사명'].str.replace('\(주\)', '')

minor.head()
```

```python
default = pd.read_csv("./drive/MyDrive/공유문서함/상장_부실사유.csv", encoding='cp949')
default
```

```python
dea_list = default["회사명"].unique().tolist()
dea_list
```

```python
# 상장 부실사유를 list로 저장
b = []
for i in dea_list:
    a = minor[minor["회사명"]==i]
    b.append(a)
```

```python
default[default["회사명"].str.contains('글로포스트')]
```

```python
# 파도타기 부실 기업들만 뽑아낸 데이터프레
default_df = pd.concat(b)
default_df
```

```python
operate = minor.drop(default_df.index)
operate["상장폐지일"] = np.NaN # 상장폐지일을 모두 nan값으로 설정
operate
```

```python
# 부실에 해당하지 않는 row들의 상폐일을 모두 nan값으로 설정한 operate 데이터프레임과 부실기업만 있는 default_df 데이터프레임 결합
all = pd.concat([default_df,operate])
all.reset_index(drop=True,inplace=True)
```
