```python
df.reset_index(inplace=True)
```

```python
df_label = df[['index','회사명', '거래소코드', '회계년도', '상장일', '상장폐지일']]
df_label
```

```python
# '회사명'이 '리드'이고 '상장폐지일'이 NaN인 행의 '상장폐지일'을 2020으로 채움
df_label.loc[df_label['회사명'] == '리드', '상장폐지일'] = df_label.loc[df_label['회사명'] == '리드', '상장폐지일'].fillna(2020)
```

```python
# '회사명'이 '오상헬스케어'이고 '상장폐지일'이 NaN인 행의 '상장폐지일'을 2016으로 채움
df_label.loc[df_label['회사명'] == '오상헬스케어', '상장폐지일'] = df_label.loc[df_label['회사명'] == '오상헬스케어', '상장폐지일'].fillna(2016)
```

```python
df_label["상폐여부"] = df_label["상장폐지일"].notna().astype(int)
df_label
```

```python
real_df = df_label[df_label['상폐여부']==1]
real_df
```

```python
df_label= real_df.groupby('회사명')['index'].apply(list)
```

```python
df_label2 = pd.DataFrame(df_label)
df_label2
```

```python
df_label2["index"][0][-2] # 데이터프레임의 index열에서 첫번째 행인 KCW의 값에서 뒤에서 두번째인 값 [161, 162, 163, 164]에서 163의미
```

```python
label_index = []
for j in df_label2.index:
    if len(df_label2["index"][j]) == 1:
        label_index.append(df_label2["index"][j][-1])
    if len(df_label2["index"][j]) >= 2:
        label_index.append(df_label2["index"][j][-1])
        label_index.append(df_label2["index"][j][-2])
```

```python
label_index
```

```python
real_df.iloc[173:177]
```

```python
# df에 부실 (t-1), (t-2)년도에 라벨링하기
for i in label_index:
    df.loc[i,'label'] = 1
df['label'] = df["label"].fillna(0)
```

```python
df['label'].value_counts()
```

```python
# 상폐일이 있는 기업만 데이터프레임으로 저장
defalut = df[~df["상장폐지일"].isnull()]
defalut
```

```python
# 상폐일이 존재하지만 부실이 0인 row의 index 찾기
drop_index=defalut[defalut['label']==0].index
drop_index
```

```python
# 부실 기업 중 레이블 1에 해당하지 않는 row는 제거
df.drop(drop_index, inplace=True)
df.drop('index',axis=1, inplace=True)
```

```python
df.reset_index(inplace=True,drop=True)
df
```
