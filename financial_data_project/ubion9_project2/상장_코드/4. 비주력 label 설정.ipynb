```python
dea_list = default["회사명"].unique().tolist()
dea_list
```

```python
# 상장 부실사유를 list로 저장
b = []
for i in dea_list:
    a = df[df["회사명"]==i]
    b.append(a)
```

```python
# 파도타기 부실 기업들만 뽑아낸 데이터프레임
default_df = pd.concat(b)
default_df
```

```python
df.reset_index(inplace=True)
```

```python
df_label = df[['index','회사명', '거래소코드', '회계년도', '상장일', '상장폐지일']]
df_label
```

```python
df_label["상폐여부"] = df_label["상장폐지일"].notna().astype(int)
df_label
```

```python
real_df= df_label[df_label['상폐여부']==1]
real_df
```

```python
df_label= real_df.groupby('회사명')['index'].apply(list)
```

```python
df_label2 = pd.DataFrame(df_label)
df_label2
```

```python
df_label2["index"][0][-2]
```

```python
label_index = []
for j in df_label2.index:
    if len(df_label2["index"][j]) == 1:
        label_index.append(df_label2["index"][j][-1])
    if len(df_label2["index"][j]) >= 2:
        label_index.append(df_label2["index"][j][-1])
        label_index.append(df_label2["index"][j][-2])
```

```python
label_index
```

```python
for i in label_index:
    df.loc[i,'label'] = 1
df['label'] = df["label"].fillna(0)
```

```python
df['label'].value_counts()
```

```python
defalut = df[~df["상장폐지일"].isnull()]
defalut
```

```python
drop_index=defalut[defalut['label']==0].index
drop_index
```

```python
df.drop(drop_index, inplace=True)
df.drop('index',axis=1, inplace=True)
```

```python
df.reset_index(inplace=True,drop=True)
df
```
