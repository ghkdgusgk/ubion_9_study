```python
 # 부동산업 기타금융 여신금융 증권 신용금고 보험업 은행업 제외
drop_list = ['부동산업','기타금융업', '여신금융업', '증권업', '신용금고업', '보험업', '은행업']
listed_ind = listed_ind[~listed_ind['상장협 산업분류(중분류)'].isin(drop_list)]
```

###### 2. 산업분류 참고하여 주력산업, 비주력산업 나누기

```python
# 상장 데이터프레임은 listed_ind
# major_list는 주력산업의 분류를 list로 저장한 것
# major_unlist는 주력산업 이외의 것들
```

```python
major_list = ['자동차 및 트레일러 제조업', '고무 및 플라스틱제품 제조업', '전자부품, 컴퓨터, 영상, 음향 및 통신장비 제조업', '화학물질 및 화학제품 제조업; 의약품 제외', '1차 금속 제조업', '전기장비 제조업', '컴퓨터 프로그래밍, 시스템 통합 및 관리업', '섬유제품 제조업; 의복제외', '의료용 물질 및 의약품 제조업', '코크스, 연탄 및 석유정제품 제조업', '의료, 정밀, 광학기기 및 시계 제조업']
major_unlist = listed_ind[listed_ind['상장협 산업분류(중분류)'].isin(major_list)]
listed_ind2 = listed_ind[~listed_ind['상장협 산업분류(중분류)'].isin(major_list)]
major_unlist.shape
```

```python
major_list2 = ['특수 목적용 기계 제조업', '일반 목적용 기계 제조업','소프트웨어 개발 및 공급업','선박 및 보트 건조업', '항공기, 우주선 및 부품 제조업']
major_unlist2 = listed_ind2[listed_ind2['상장협 산업분류(소분류)'].isin(major_list2)]
```

```python
main_industry =pd.concat([major_unlist, major_unlist2], axis=0)
non_main_industry = listed_ind2[~listed_ind2['상장협 산업분류(소분류)'].isin(major_list2)]
```
