```python
# 연결개별 합치기
df_1 = []
df_2 = []
for i in ma_flow["회사명"].unique():
    a = ma_flow[ma_flow["회사명"]==i].reset_index(drop=True)
    b = ma_unflow[ma_unflow["회사명"]==i].reset_index(drop=True)
    if len(a) == len(b): # "ma_flow"와 "ma_unflow"에서 선택된 데이터의 길이가 같다면
        for j in feature:
            a[j] = a[j].fillna(b[j]) # "ma_flow" 데이터프레임의 특정 피처 값이 결측치라면, 해당 결측치를 "ma_unflow" 데이터프레임의 동일한 피처 값으로 채움
        df_1.append(a)
    else:
        df_2.append(a)
```

```python
# 데이터프레임으로 변경
df_22 = pd.concat(df_2)
df_11= pd.concat(df_1)
df_22
```
