### 데이터프레임의 결합 예제
1. csv폴더에서 tran_1.csv, tran_2.csv 파일 로드 
2. 1번과정에서 나오는 2개의 데이터프레임을 유니언 결합
3. csv 폴더에서 tran_d_1.csv, tran_d_2.csv 파일 로드 
4. 3번 과정에서 나오는 2개의 데이터프레임을 유니언 결합
5. 2번 과정에서 나온 데이터프레임과 4번 과정 나온 데이터프레임을 특정 조건에 맞춰서 조인 결합 (left)
6. 5번 과정에서 나온 데이터프레임과 csv 폴더 안에 있는 item_master.csv파일을 로드하여 2개의 데이터프레임을 조인 결합

```python
import pandas as pd
```

```python
tran_1 = pd.read_csv('../../csv/tran_1.csv')
tran_2 = pd.read_csv('../../csv/tran_2.csv')
tran_d_1 = pd.read_csv('../../csv/tran_d_1.csv')
tran_d_2 = pd.read_csv('../../csv/tran_d_2.csv')
item_master = pd.read_csv('../../csv/item_master.csv')
```

```python
# tran_1, tran_2 유니언 결합
tran = pd.concat([tran_1, tran_2], axis=0, ignore_index=True)
```

```python
# tran_d_1, tran_d_2 유니언 결합 
tran_d = pd.concat([tran_d_1, tran_d_2], axis=0, ignore_index=True)
```

```python
# tran 데이터의 transaction_id 값들의 빈도수 
tran['transaction_id'].value_counts()
```

```python
tran_d['transaction_id'].value_counts()
```

```python
tran.head(3)
```

```python
tran_d.head(3)
```

```python
# tran, tran_d 조인결합 
transaction = pd.merge(tran, tran_d, on='transaction_id', how='left')
```

```python
transaction['transaction_id'].value_counts()
```

```python
transaction.head(2)
```

```python
item_master
```

```python
total_df = pd.merge(transaction, item_master, on='item_id', how='left')
```

```python
total_df.sort_values(total_df.columns[0])
```

```python
# total_df에서 가장 많이 팔린 제품의 이름은 무엇인가?
total_df.head()
```

```python
group_data = total_df[ ['item_name', 'quantity'] ].groupby('item_name').sum()
```

```python
group_data.sort_values('quantity', ascending=False)
```

```python
# 가장 매출이 높은 아이템의 이름은 무엇인가?
total_df['total_price'] = total_df['quantity'] * total_df['item_price']
```
