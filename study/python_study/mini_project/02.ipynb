```python
# 2)yf.download 사용 
# 코스피는 KS, 코스닥은 KQ

hanhwa_air = yf.download("012450.KS")
hanhwa_air = hanhwa_air.tail(500)
```

```python
# 한화에어로스페이스 (단순)이동평균선 생성
hanhwa_air['SMA_5'] =talib.SMA(hanhwa_air['Close'], 5)
hanhwa_air['SMA_20'] =talib.SMA(hanhwa_air['Close'], 20)
hanhwa_air['SMA_60'] =talib.SMA(hanhwa_air['Close'], 60)
hanhwa_air['SMA_120'] =talib.SMA(hanhwa_air['Close'], 120)
hanhwa_air['SMA_240'] =talib.SMA(hanhwa_air['Close'], 240)
```

```python
# 차트화
hanhwa_air[['Close', 'SMA_5', 'SMA_20', 'SMA_60', 'SMA_120', 'SMA_240']].plot(figsize=(14,6))
```
