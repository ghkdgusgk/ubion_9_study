```python
import matplotlib.pyplot as plt
# from matplotlib import pyplot as plt
```

```python
# 라인 그래프 생성 (y축 데이터만 존재하는 경우)
plt.plot([5, 2, 9, 1])
plt.show()
```

```python
# x축 데이터와 y축 데이터가 모두 존재하는 경우 라인 그래프
x = [1, 2, 3, 4]
y = [10, 5, 2, 9]

plt.plot(x, y)
plt.show()
```

```python
# 라인 그래프에 옵션 (색상, 선의 종류)
plt.plot(x, y, '#FF0066')
plt.show()
```

```python
plt.plot(x, y, '--g')
plt.show()
```

```python
plt.plot(x, y)
plt.plot(x, [30, 20, 10, 20])
plt.show()
```

```python
# x축의 이름을 지정, y축의 이름을 지정, 그래프의 이름을 지정 
plt.plot(x, y)
plt.xlabel('A')
plt.ylabel('B')
plt.title('Graph')
plt.show()
```

```python
# 범례 추가
plt.plot(x, y)
plt.plot(x, [5, 10, 15, 20])
plt.legend(['a', 'b'], loc='upper right')
plt.show()
```

```python
# 바형 그래프 
x = [1,2,3,4]
y = [80, 70, 90, 95]

plt.bar(x, y)
plt.show()
```

```python
plt.barh(x, y)
plt.show()
```

```python
# 옵션 추가 
plt.bar(x, y, color = ['r', 'g','b'], width=0.3)
plt.show()
```

```python
import seaborn as sns 
import pandas as pd
```

```python
drinks = pd.read_csv('../../csv/drinks.csv')
```

```python
drinks.head(3)
```

```python
corr = drinks[
    ['beer_servings', 'spirit_servings', 
     'wine_servings', 'total_litres_of_pure_alcohol']
].corr(method='pearson')
corr
```

```python
# 히트맵 출력
cols_view = ['beer', 'spirit', 'wine', 'alcohol']

hm = sns.heatmap(
    corr.values, 
    cbar = True, # 사이드바
    annot = True, # 히트맵에서 숫자를 표시 
    square= True,  # 히트맵의 정사각형 표시 
    fmt = '.3f', #히트맵의 숫자의 소수점 자리수
    annot_kws= {'size' : 20}, # 숫자의 크기
    xticklabels= cols_view, 
    yticklabels= cols_view
)

plt.tight_layout()
plt.show()
```

```python
sns.pairplot(
    drinks[[
        'beer_servings', 
        'spirit_servings', 
        'wine_servings', 
        'total_litres_of_pure_alcohol'
    ]], 
    height=2.5
)
plt.show()
```
